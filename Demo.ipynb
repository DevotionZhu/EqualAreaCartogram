```python
from eqcart import Cartogram
```

### Demo 1: Converting Assembly Constituencies in the state of Gujarat, India into an Equal Area Hexagonal Choropleth

##### Input File

We will use a geojson file as our input. This is what it looks like

```python
%matplotlib inline
import geopandas as gdf
temp_df = gdf.read_file("sample_data/gujarat.json")
temp_df.plot()
```

##### First attempt

```python
cart = Cartogram("sample_data/gujarat.json", #input filepath
                 id_col="AC_NO", num_x_grid=50, num_y_grid=40)
cart.make_hex_svg("sample_output/gujarat_sparse.svg", show=True)
```

This is fairly accurate. But the map might be a little too sparse for your taste. Let's try to make it more dense.

##### Second attempt

```python
cart = Cartogram("sample_data/gujarat.json", #input filepath
                 id_col="AC_NO", 
                 num_x_grid=25, num_y_grid=20 #reduced grid width and height to make map more dense
                )
cart.make_hex_svg("sample_output/gujarat_dense.svg", show=True)
```

Much more dense. We could use Inkscape or Illustrator to bring individual areas closer together or farther apart.

##### Saving as GeoJSON

```python
cart = Cartogram("sample_data/gujarat.json", #input filepath
                 id_col="AC_NO", 
                 num_x_grid=25, num_y_grid=20 #reduced grid width and height to make map more dense
                )
cart.make_hex_geojson("sample_output/gujarat_dense.geojson")
```

```python
gdf.read_file("sample_output/gujarat_dense.geojson").plot()
```

### Demo 2: Converting Assembly Constituencies in the state of Himachal Pradesh, India into an Equal Area Hexagonal Choropleth

This is what our input file looks like

```python
%matplotlib inline
import geopandas as gdf
gdf.read_file("sample_data/himachal.json").plot()
```

##### Attempt 1

```python
cart = Cartogram("sample_data/himachal.json", id_col="id", num_x_grid=10, num_y_grid=10)
cart.make_hex_svg("sample_output/himachal_dense.svg", show=True)
```

This works, but let's try a slightly more sparse map for comparison

```python
cart = Cartogram("sample_data/himachal.json", id_col="id", num_x_grid=15, num_y_grid=15)
cart.make_hex_svg("sample_output/himachal_sparse.svg", show=True)
```

This captures the geography of the area somewhat better
